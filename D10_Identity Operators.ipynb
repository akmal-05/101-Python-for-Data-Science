<h1>Identity Operators</h1>


Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location:

is Operator

Description: Returns True if both variables refer to the same object in memory.
It checks identity, not just value.

```python
x = ["apple", "banana"]
y = x
print(x is y)    # Output: True
```

is not Operator

Description: Returns True if both variables are not the same object (different memory locations).

```python
x = ["apple", "banana"]
y = ["apple", "banana"]
print(x is not y)   # Output: True
```

```python
x = [1, 2, 3]
y = x
print(x is y)    # True
```

```python
y.append(4)
print(x)  # Output: [1, 2, 3, 4]
print(x is y) 
```

```python
x = [1, 2, 3]
y = x
print(x == y)    # True  (values are equal)
print(x is y)    # False (different objects)
print(x is not y) # True  (different objects)
```

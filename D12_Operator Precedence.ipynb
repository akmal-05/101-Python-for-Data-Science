<h1>Operator Precedence
</h1>
operator precedence describes the order in which operations are performed.


```python
x = 2 + 3 * 4
print(x)
```

```python
x = (2 + 3) * 4
print(x)
```

```python
x = 10 + 6 / 2
print(x)
```

```python
x = 2 ** 3 ** 2
print(x)
```

```python
x = 10 // 3 + 10 % 3
print(x)
```

```python
x = not 10 > 5
print(x)
```

```python
x = 5 and 0 or 4
print(x)
```

```python
x = (5 > 3) and (2 < 1)
print(x)
```

```python
x = (5 > 3) or (2 < 1)
print(x)
```

```python
x = 4 + 3 * 2 ** 2
print(x)
```

```python
x = (10 - 4) * 2 + 3 ** 2
print(x)
```

```python
x = 4 + 5 * 2
print(x)
```

```python
x = (4 + 5) * 2
print(x)
```

```python
x = 2 ** 3 ** 2
print(x)
```

```python
x = 15 // 4 + 15 % 4
print(x)
```

```python
x = not (10 < 5)
print(x)
```

```python
x = 5 > 3 and 2 < 1
print(x)
```

```python
x = (3 + 2) * 2 ** 3
print(x)
```

```python
x = 5 and 0 or 7
print(x)
```

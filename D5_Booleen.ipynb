<h1>Python Booleans</h1>
Booleans represent one of two values: True or False.

Boolean Values
1) In programming, you often need to know if an expression is True or False.
2) You can evaluate any expression in Python and get one of two answers:  True or False.
3) When you compare two values, the expression is evaluated and Python returns the Boolean answer:


```python
print(10 > 9)
print(10 == 9)
print(10 < 9)
```

When you run a condition in an if statement, Python returns True or False:


```python
a = 200
b = 33222

if b > a:
  print("b is greater than a")
  print("This is inside if block")
else:
  print("b is not greater than a")
```



In Python, any non-zero number (positive or negative) is considered True, while 0 is False.

```python
a = 200
print(bool(a))
```

In Python, any non-empty string is True, and an empty string ("") is False.

```python
b="Hello"
print(bool(b))
```

```python
print(bool(0))      # False, because 0 means "nothing"
print(bool(15))     # True, any non-zero number is True
print(bool(-10))    # True, negative numbers are also True
print(bool(0.0))    # False, same as 0
print(bool(3.14))   # True
```

```python
print(bool(""))         # False, empty string
print(bool(" "))        # True, space counts as a character
print(bool("Python"))   # True, non-empty string
print(bool([]))         # False, empty list
```

<h1>Python Comparison Operators</h1>

Comparison operators are used to compare two values:



== (Equal To)
Returns True if both values are equal.

```python
x = 5
y = 5
print(x == y)   # Output: True
```

!= (Not Equal To)

Definition: Returns True if the values are not equal.

```python
x = 5
y = 3
print(x != y)   # Output: True
```

> (Greater Than)

Definition: Returns True if the left value is greater than the right value.

```python
x = 5
y = 3
print(x > y)    # Output: True
```

< (Less Than)

Definition: Returns True if the left value is less than the right value.

```python
x = 5
y = 8
print(x < y)    # Output: True
```

>= (Greater Than or Equal To)

Definition: Returns True if the left value is greater than or equal to the right value.

```python
x = 5
y = 5
print(x >= y)   # Output: True
```

<= (Less Than or Equal To)

Definition: Returns True if the left value is less than or equal to the right value.

```python
x = 3
y = 5
print(x <= y)   # Output: True
```

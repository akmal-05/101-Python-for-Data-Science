 Python Collections (Arrays)

There are four collection data types in the Python programming language:
1. A list is a collection that is ordered and changeable. Allows duplicate members.
2. A tuple is a collection that is ordered and unchangeable. Allows duplicate members.
3. A set is a collection that is unordered, unchangeable*, and unindexed. No duplicate members.
4. A dictionary is a collection that is ordered and changeable. No duplicate members.


<h1>Python Lists</h1>
In Python, a list is an ordered collection of items that can hold multiple values in a single variable. Lists are mutable, meaning you can change, add, or remove elements after creating them.
Lists are one of 4 built-in data types in Python used to store collections of data; the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.

```python
fruits=["Mango","Appple","Banana"]
fruits
```

```python
my_list = [10, 20, 30, 40]
print(my_list)
```

1. List items are ordered, changeable, and allow duplicate values.
2. List items are indexed; the first item has index [0], the second item has index [1], etc.
3. When we say that lists are ordered, it means that the items have a defined order, and that order will not change.
4. If you add new items to a list, they will be placed at the end of the list.
5. Some list methods will alter the order, but generally, the order of the items remains unchanged
6. The list is changeable, meaning that we can change, add, and remove items in a list after it has been created.


Allow Duplicates
Since lists are indexed, lists can have items with the same value:


```python
my_list = [10, 20, 10,30, 40]
my_list
```

List Length
To determine how many items a list has, use the len() function:

```python
thislist = ["apple", "banana", "cherry", "apple", "cherry"]
print(len(thislist))
```

<h3>Python - Access List Items</h3>
In Python, you can access list items using their index numbers. Each item in a list has a position, starting from index 0 for the first item.

<h4>1. Accessing by Index</h4>


```python
fruits = ["apple", "banana", "cherry"]
print(fruits[1])
```

<h4>2. Negative Indexing</h4>

Negative indexing means starting from the end

-1 refers to the last item, -2 refers to the second last item, etc.


```python
fruits = ["apple", "banana", "cherry"]
print(fruits[-3])
```

<h4>3. Range of Indexes</h4>
You can specify a range of indexes by specifying where to start and where to end the range.
When specifying a range, the return value will be a new list with the specified items.


```python
fruits = ["apple", "banana", "cherry", "orange", "kiwi"]
print(fruits[1:4])
```

By leaving out the start value, the range will start at the first item:

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:4])
```

By leaving out the end value, the range will go on to the end of the list:

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:])
```

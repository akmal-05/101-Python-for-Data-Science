 Python Collections (Arrays)

There are four collection data types in the Python programming language:
1. A list is a collection that is ordered and changeable. Allows duplicate members.
2. A tuple is a collection that is ordered and unchangeable. Allows duplicate members.
3. A set is a collection that is unordered, unchangeable*, and unindexed. No duplicate members.
4. A dictionary is a collection that is ordered and changeable. No duplicate members.


<h1>Python Lists</h1>
In Python, a list is an ordered collection of items that can hold multiple values in a single variable. Lists are mutable, meaning you can change, add, or remove elements after creating them.
Lists are one of 4 built-in data types in Python used to store collections of data; the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.

```python
fruits=["Mango","Appple","Banana"]
fruits
```

```python
my_list = [10, 20, 30, 40]
print(my_list)
```

1. List items are ordered, changeable, and allow duplicate values.
2. List items are indexed; the first item has index [0], the second item has index [1], etc.
3. When we say that lists are ordered, it means that the items have a defined order, and that order will not change.
4. If you add new items to a list, they will be placed at the end of the list.
5. Some list methods will alter the order, but generally, the order of the items remains unchanged
6. The list is changeable, meaning that we can change, add, and remove items in a list after it has been created.

<h3>Python - Access List Items</h3>
In Python, you can access list items using their index numbers. Each item in a list has a position, starting from index 0 for the first item.

<h4>1. Accessing by Index</h4>


```python
fruits = ["apple", "banana", "cherry"]
print(fruits[1])
```

<h4>2. Negative Indexing</h4>

Negative indexing means starting from the end

-1 refers to the last item, -2 refers to the second last item, etc.


```python
fruits = ["apple", "banana", "cherry"]
print(fruits[-3])
```

<h4>3. Range of Indexes</h4>
You can specify a range of indexes by specifying where to start and where to end the range.
When specifying a range, the return value will be a new list with the specified items.


```python
fruits = ["apple", "banana", "cherry", "orange", "kiwi"]
print(fruits[1:4])
```

By leaving out the start value, the range will start at the first item:

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:4])
```

By leaving out the end value, the range will go on to the end of the list:

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:])
```

Range of Negative Indexes
Specify negative indexes if you want to start the search from the end of the list:


```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[-4:-1])
```

<h3>Python - Change List Items</h3>
In Python, lists are mutable, meaning you can change, replace, or update their elements after creation.


<h4>1. Change a Specific Item</h4>
Use the index number to modify a single value.


```python
fruits = ["apple", "banana", "cherry"]
fruits[2] = "mango"
print(fruits)
```

<h4>2. Change a Range of Items</h4>
You can replace multiple items at once using slicing.


```python
fruits = ["apple", "banana", "cherry", "orange", "kiwi"]
fruits[1:3] = ["grape", "melon"]
print(fruits)
```

<h4>3. Replace Multiple Items with Fewer or More</h4>
When you assign a different number of values to a slice, the list adjusts automatically.


```python
fruits = ["apple", "banana", "cherry", "orange"]
fruits[1:3] = ["mango"]
print(fruits)
```

<h3>Python - Add List Items</h3>
In Python, you can add new items to a list in several ways since lists are mutable. You can use methods like append(), insert(), or extend() depending on where and how you want to add elements.


<h4>1. Add an Item to the End – append()</h4>
Adds a single element to the end of the list.


```python
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
print(fruits)
```

<h4>2. Add an Item at a Specific Position – insert()<h4>
Adds an item at a specific index.


```python
fruits = ["apple", "banana", "cherry"]
fruits.insert(1, "mango")
print(fruits)
```

<h4>3. Add Multiple Items – extend()</h4>
Adds all elements from another list (or iterable) to the current list.


```python
fruits = ["apple", "banana", "cherry"]
more_fruits = ["orange", "kiwi"]
fruits.extend(more_fruits)
print(fruits)
```

<h4>4. Add Items Using + Operator<h4>
You can combine two lists using the + operator.

```python
a = [1, 2, 3]
b = [4, 5]
c = a + b
print(c)
```

```python
fruits = ["apple", "banana", "cherry"]
more_fruits = ["orange", "kiwi"]
c=fruits+more_fruits
print(c)
```

<h3>Remove Specified Item</h3>
In Python, you can remove specific items from a list using several methods, depending on what you want to remove and how.

<h4>1. Remove by Value – remove()</h4>
Removes the first occurrence of a specified item

```python
fruits = ["apple", "banana", "cherry", "banana"]
fruits.remove("banana")
print(fruits)
# If there is more than one item with the specified value, the remove() method removes the first occurrence
```

<h4>2. Remove by Index – pop()</h4>
Removes an item by its index.
 If no index is given, it removes the last item by default.

```python
fruits = ["apple", "banana", "cherry"]
fruits.pop(1)
print(fruits)
```

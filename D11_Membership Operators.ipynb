<h1>Membership Operators</h1>
Membership operators in Python are used to check whether a value or variable is present in a sequence such as a list, tuple, string, or set.

in Operator

Description: Returns True if the specified value exists in the sequence (like list, tuple, string, or set).

```python
fruits = ["apple", "banana", "cherry"]
print("banana" in fruits)   # Output: True
```

not in Operator

Description: Returns True if the specified value does not exist in the sequence.

```python
fruits = ["apple", "banana", "cherry"]
print("mango" not in fruits)   # Output: True
```

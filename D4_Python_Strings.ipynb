# Strings
Strings in Python are surrounded by either single quotation marks or double quotation marks.
'hello' is the same as "hello".
You can display a string literal with the print() function:

```python
name='User'
print(f"{name} Greets you!")
# print('Hello')
```

### Quotes Inside Quotes
You can use quotes inside a string, as long as they don't match the quotes surrounding the string:


```python
print("Its fine dear")
print("Its fine dear'Akma'")
print('its fine dear "akma"')
```

### Assign String to a Variable
Assigning a string to a variable is done with the variable name followed by an equal sign and the string:


```python
c=("This is a string to check its length and type")
print(len(c))
print(type(c))
```

### Multiline Strings
You can assign a multiline string to a variable by using three quotes:


```python
a="""This is a multi line string
to check how multi line strings"""
print(a)
print(type(a))
```

### Strings are Arrays
Like many other popular programming languages, strings in Python are arrays of unicode characters.
However, Python does not have a character data type; a single character is simply a string with a length of 1.
Square brackets can be used to access elements of the string.


```python
a="SZABIST"
print(a[4])
print(type(a[4]))
 
print(a[-3])
```

### Looping Through a String
Since strings are arrays, we can loop through the characters in a string, with a for loop.

```python
fruts = "banana"
for x in fruts:
  print(x)
```

### String Length
To get the length of a string, use the len() function.

```python
strr="This is a sentence"
print(len(strr))
```

### Python - Slicing Strings
In Python, string slicing means extracting a part (substring) of a string by specifying a range of indexes. In short, slicing gives you powerful control to pick portions of strings without writing loops.

```python
b = "Hello, World!"
print(b[3:8])
```

```python
text = "Python Programming"
part = text[0 : 6]
print(part)
print(part.upper())
```

```python
c="This is a sentence"
print(c[3:11])
```

### Slice From the Start
By leaving out the start index, the range will start at the first character:


```python
b = "Hello, World!"
print(b[:])
```

```python
d="Python is a programming language"
print(d[:8])
```

### Slice To the End
By leaving out the end index, the range will go to the end:


```python
b = "Hello, World!"
print(b[2:])
```

```python
d="Python is a programming language"
print(d[4:])
```

### Negative Indexing
Use negative indexes to start the slice from the end of the string:


```python
b = "Hello, World!"
print(b[-5:-2])
```

```python

d="Python is a programming language"
print(d[-9:-3])
```

### Python - Modify Strings
Python has a set of built-in methods that you can use on strings.


```python
# The upper() method returns the string in upper case:
a = "Hello, World!"
print(a.upper())
```

```python
# The lower() method returns the string in lower case:
a = "Hello, World!"
print(a.lower())
```

```python
# The strip() method removes any whitespace from the beginning or the end:
a = " Hello, World! "
print(a.strip()) # returns "Hello, World!"
```

```python
# The replace() method replaces a string with another string:
a = "Hello, World!"
print(a.replace("H", "J"))
```

```python
# The split() method splits the string into substrings if it finds instances of the separator:
a = "Hello, World!"
print(a.split(",")) # returns ['Hello', ' World!']
```

### Python - String Concatenation
In Python, string concatenation means joining two or more strings into a single string. To concatenate, or combine, two strings, we can use the + operator.


```python
# Merge variable a with variable b into variable c:
a = "Hello"
b = "World"
c = a+b
print(c)
```

```python
# To add a space between them, add a " ":
a = "Hello"
b = "World"
c = a + " " + b
print(c)
```

### Python - Format - Strings

```python
# As we learned in the Python Variables chapter, we cannot combine strings and numbers like this:
age = 36
#This will produce an error:
txt = "My name is John, I am " + age
print(txt)
# But we can combine strings and numbers by using f-strings or the format() method!
```

### F-Strings
F-String was introduced in Python 3.6, and is now the preferred way of formatting strings.
To specify a string as an f-string, simply put an f in front of the string literal, and add curly brackets {} as placeholders for variables and other operations.


```python
age = 36
txt = f"My name is John, I am {age}"
print(txt)
```

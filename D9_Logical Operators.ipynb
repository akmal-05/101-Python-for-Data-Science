<h1>Python Logical Operators</h1>
Logical operators are used to combine conditional statements:


and Operator

Description: Returns True if both conditions are true.

```python
x = 7
print(x > 5 and x < 10)   # Both conditions true → Output: True
```

or Operator

Description: Returns True if at least one of the conditions is true.

```python
x = 7
print(x < 5 or x < 10)    # One condition true → Output: True
```

not Operator

Description: Reverses the result, returns False if the condition is true.

```python
x = 7
print(not(x > 5 and x < 10))   # Inside is True, so result becomes False
```

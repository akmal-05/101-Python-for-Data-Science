<h1>Data type</h1>

In simple terms, a data type tells Python what kind of value you are working with, so the computer knows how to store it and what operations are allowed on it.
 

1. Integer (int)

Whole numbers without decimal points.

```python
age = 21
year = 2025
temperature = -5
print(age, year, temperature)
```

```python
a = 77
print(a)

kamran=33
print(type(kamran))
type(kamran)
```

2. Float (float)

Numbers with decimals.

```python
pi = 3.14
weight = 65.7
temperature = -2.5
print(pi, weight, temperature)
```

```python
percentage =88.9
radius=33.0
print(percentage,radius)

type(radius)
```

3. String (str)

Text inside quotes.

```python
name = "Akmal"
greeting = 'Hello Data Science'
combo= name   + greeting
print(combo)
```

```python
text="This is my heading"
print(text)
```

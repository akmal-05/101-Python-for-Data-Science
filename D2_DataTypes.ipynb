
<h1>Variables</h1>

Variable is like a container that holds data. Very similar to how our containers in kitchen holds sugar, salt etc Creating a variable is like creating a placeholder in memory and assigning it some value.

```python
a= 1234567
print(a)
# print(type(a))

b= 'harry'
print(b)
type(b)

c= True
print(c)

d= 66.3
print(type(d))
```

<h1>Data type</h1>

In simple terms, a data type tells Python what kind of value you are working with, so the computer knows how to store it and what operations are allowed on it.
 

1. Integer (int)

Whole numbers without decimal points.

```python
age = 21
year = 2025
temperature = -5
print("these are variabes  ",age, year, temperature)
```

```python
a = 77
print(a)

kamran=33
print(type(kamran))
type(kamran)

b= ('harry')
c= 'ijaz'
d=b+c
print(d)
```

2. Float (float)

```python
pi = 3.14
weight = 65.7
temperature = -2.5
print(pi, weight, temperature)
```

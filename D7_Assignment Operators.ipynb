<h1>Assignment Operators</h1>

Assignment operators are used to assign values to variables:

= (Assignment Operator)

```python
x = 5
print(x)   # Output: 5
```

+= (Add and Assign)

```python
x = 5
x += 3    # same as x = x + 3
print(x)  # Output: 8
```

-= (Subtract and Assign)

```python
x = 5
x -= 3    # same as x = x - 3
print(x)  # Output: 2
```

*= (Multiply and Assign)

```python
x = 5
x *= 3    # same as x = x * 3
print(x)  # Output: 15
```

/= (Divide and Assign)

```python
x = 5
x /= 3    # same as x = x / 3
print(x)  # Output: 1.6666666666666667
```

%= (Modulus and Assign)

```python
x = 5
x %= 3    # same as x = x % 3
print(x)  # Output: 2
```

//= (Floor Divide and Assign)

```python
x = 5
x //= 3   # same as x = x // 3
print(x)  # Output: 1
```

**= (Exponentiate and Assign)

```python
x = 5
x **= 3   # same as x = x ** 3
print(x)  # Output: 125
```

&= (Bitwise AND and Assign)


```python
x = 5
x &= 3    # same as x = x & 3
print(x)  # Output: 1
```

|= (Bitwise OR and Assign)

```python
x = 5
x |= 3    # same as x = x | 3
print(x)  # Output: 7
```

^= (Bitwise XOR and Assign)

```python
x = 5
x ^= 3    # same as x = x ^ 3
print(x)  # Output: 6
```

>>= (Right Shift and Assign)

```python
x = 8
x >>= 2   # same as x = x >> 2
print(x)  # Output: 2
```

<<= (Left Shift and Assign)

```python
x = 3
x <<= 2   # same as x = x << 2
print(x)  # Output: 12
```

:= (Walrus Operator)

```python
print(x := 3)  # Assigns 3 to x and prints 3
print(x)       # Output: 3
```
